```python
light = input("Enter the colour:")
if(light == "red"):
    print("Please stop.")
elif (light == "green"):
    print("Please go.")
elif (light == "yellow"):
    print("Please wait.")
else:
    print("Invalid colour.")
```

```python
marks = int(input("Enter the marks you secured:"))
if(marks>=90):
    print("Distinction with Grade A")
elif (marks>=75 and marks<90):
    print("First class with Grade B")
elif (marks>=50 and marks<75):
    print("Second class with Grade C")
elif (marks<50):
    print("Fail,Please try again next year")
else:
    print("Please enter correct value")
    
```

```python
no_consumed = float(input("Enter the units consumed:"))
if(no_consumed<=100):
    print("the amt to be paid is:",0.5*100)
elif (no_consumed<=200):
    print("The amt to be paid is:",(0.5*100)+((no_consumed-100)*0.75))
elif (no_consumed>200):
    print("The amt to be paid is:",(0.5*100)+(100*0.75)+((no_consumed-200)*1))
```
