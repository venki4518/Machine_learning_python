```python
length = float(input("Enter the side of the square:"))
Radius = float(input("Enter the radius of the circle:"))
sqarea = length**2
crarea = (3.14*((Radius)**2))
print(f"The area of square is {sqarea} and area of circle is {crarea}")
```

```python
length = float(input("Enter the side of the square:"))
sqperi = length*4
print(f"The perimeter of the square of side {length} is = {sqperi}")
```

```python
Height = float(input("Enter the height of the triangle:"))
Base = float(input("Enter the base of the triangle:"))
triarea = 0.5*Height*Base
print(f"The area of triangle is = {triarea}")
```

```python
Radius = float(input("Enter the radius of the circle:"))
circum = 2*3.14*Radius
print(f"The circumferance of the circle is = {circum}")
```

```python
principle = float(input("Enter the principle amount:"))
Rate = float(input("Enter the Rate of interest:"))
Time = float(input("Enter the time/period of interest:"))
simple_Interest = (principle*Rate*Time)/100
print(f"The simple interest is = {simple_Interest}")
```
